# Sales Analysis

#### Import Necessary Libraries

```python
import pandas as pd
import os
```

#### Merging 12 months of data in a single data file

```python
files = [file for file in os.listdir('./Sales_Data/')]
all_months_data = pd.DataFrame()

for file in files:
    df = pd.read_csv('./Sales_Data/'+file)
    all_months_data = pd.concat([all_months_data, df])

all_months_data.to_csv("all_data.csv",index=False)
```

#### Read in updated dataframe

```python
all_data = pd.read_csv('./all_data.csv')
all_data.head()
```

### Clean up the data!

#### Drop rows of NaN

```python
#nan_df = all_data[all_data.isna().any(axis=1)]  # we just check where is NaN
all_data = all_data.dropna(how='all')           # we remove all NAN data using 'dropna'
```

```python
all_data = all_data[all_data['Order Date'].str[0:2] != "Or"]
all_data.head()
```

##### Convert columns to correct data types

```python
all_data['Quantity Ordered'] = pd.to_numeric(all_data['Quantity Ordered'])# convert to int/ number
all_data['Price Each'] = pd.to_numeric(all_data['Price Each'])# convert to float/ number
```

### Add Month Column

```python
all_data['Month'] = all_data['Order Date'].str[0:2]
all_data['Month'] = all_data['Month'].astype('int32')
all_data.head()
```

### Add a sales column

```python
all_data['Sales'] = all_data['Quantity Ordered'] * all_data['Price Each']
all_data.head()
```

#### Add a City column

```python
def get_city(address):
    return address.split(',')[1]

def get_state(address):
    return address.split(',')[2].split(' ')[1]

all_data['City'] = all_data['Purchase Address'].apply(lambda x: f"{get_city(x)} ({get_state(x)})")
all_data.head()
```

#### 1: What was the best month for sales? How much was earned that month? 

```python
results = all_data.groupby('Month')[['Quantity Ordered','Sales']].sum()
```

```python
import matplotlib.pyplot as plt

months = range(1,13)

plt.bar(months, results['Sales'])
plt.xticks(months)
plt.ylabel("Sales in USD ($)")
plt.xlabel("Months Number")
plt.show()
```

#### 2: What city has the highest number of sales

```python
results = all_data.groupby("City").sum()
```

```python
import matplotlib.pyplot as plt
cities = [city for city, df in all_data.groupby('City')]

plt.bar(cities, results['Sales'])
plt.xticks(cities, rotation ='vertical', size=8)
plt.ylabel("Sales in USD ($)")
plt.xlabel("City")
plt.show()
```

#### 3: What time should we display advertisement to maximize likelihood of customer's buying product?

```python
all_data.head()
```

```python
all_data['Order Date'] = pd.to_datetime(all_data['Order Date'])
all_data.head()
```

```python
all_data['Hour'] = all_data['Order Date'].dt.hour
all_data['Minute'] = all_data['Order Date'].dt.minute
all_data.head()
```

```python
hours = [hour for hour, df in all_data.groupby('Hour')]

plt.plot(hours, all_data.groupby(['Hour']).count())
plt.xticks(hours)
plt.xlabel("Hours")
plt.ylabel("Number of Orders")
plt.grid()
plt.show()

# My recommendation is around 11.am (11) or 7.pm (19)
```

#### 4: What products are often sold together?

```python
all_data.head()
```

```python
df = all_data[all_data['Order ID'].duplicated(keep=False)]

df['Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))

df = df[['Order ID','Grouped']].drop_duplicates()

df.head()
```

```python
from itertools import combinations
from collections import Counter

count = Counter()

for row in df['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list,2)))

for key, value in count.most_common(10):
    print(f"{key} - {value}")
```
